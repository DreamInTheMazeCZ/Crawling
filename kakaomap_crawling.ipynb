```python
from selenium.webdriver.common.keys import Keys

input_text = '부천시 심리상담'
driver.find_element(By.XPATH, '//*[@id="search.keyword.query"]').send_keys(input_text)

try:
    driver.find_element(By.XPATH, '//*[@id="search.keyword.submit"]').click()
    driver.implicitly_wait(20)
except:
    driver.find_element(By.XPATH, '//*[@id="search.keyword.submit"]').send_keys(Keys.ENTER)
    driver.implicitly_wait(20)
```

```python
import time

name, address, tel_num, page_url = [], [], [] ,[]
cnt = 0

while True:
    for page_num in range(1, 6):
        try:           
            # 한 페이지 당 15개 출력
            for i in range(1, 16):
                name.append(driver.find_element(By.XPATH, f'//*[@id="info.search.place.list"]/li[{i}]/div[3]/strong/a[2]').text)
                address.append(driver.find_element(By.XPATH, f'//*[@id="info.search.place.list"]/li[{i}]/div[5]/div[2]/p[1]').text)
                try:
                    tel_num.append(driver.find_element(By.XPATH, f'//*[@id="info.search.place.list"]/li[{i}]/div[5]/div[4]/span[1]').text)
                except:
                    tel_num.append('')

                try:
                    page_url.append(driver.find_element(By.XPATH, f'//*[@id="info.search.place.list"]/li[{i}]/div[5]/div[4]/a[2]').get_attribute('href'))
                except:
                    page_url.append('')
                time.sleep(1)

            if page_num == 1:
                try:
                    driver.find_element(By.XPATH, '//*[@id="info.search.place.more"]').click()
                    driver.implicitly_wait(20)
                except:
                    try:
                        driver.find_element(By.XPATH, '//*[@id="info.search.place.more"]').send_keys(Keys.ENTER)
                        driver.implicitly_wait(20)
                    except:
                        driver.find_element(By.XPATH, f'//*[@id="info.search.page.no{page_num+1}"]').click()
                time.sleep(3)

            else:
                if page_num == 5:
                    try:
                        driver.find_element(By.XPATH, ' //*[@id="info.search.page.next"]').click()
                        driver.implicitly_wait(20)
                    except:
                        driver.find_element(By.XPATH, ' //*[@id="info.search.page.next"]').send_keys(Keys.ENTER)
                        driver.implicitly_wait(20)
                    page_num += 1
                    time.sleep(3)

                else:
                    try:
                        driver.find_element(By.XPATH, f'//*[@id="info.search.page.no{page_num+1}"]').click()
                        driver.implicitly_wait(20)
                    except:
                        driver.find_element(By.XPATH, f'//*[@id="info.search.page.no{page_num+1}"]').send_keys(Keys.ENTER)
                        driver.implicitly_wait(20)
                    time.sleep(3)

        except Exception as e:
            print(e)
            cnt += 1
            break
    if cnt == 2:
        break
```
